Создаем новую колонку daytime, используя функцию Pandas Cut(). Она используется для разделения элементов массива в отдельные контейнеры и работает только на одномерных объектах, подобных массивам.
Используемые параметры:
df['hour'] - это входной массив. 
bins - определяет края бин для сегментации.
labels - используется для уточнения меток для возврата bins.

```python
df['daytime'] = pd.cut(df['hour'],
                            bins=[-1, 3, 6, 10, 16, 19, 24],
                            labels=['night', 'early morning', 'morning', 'afternoon', 'early evening', 'evening']
                           )
```
